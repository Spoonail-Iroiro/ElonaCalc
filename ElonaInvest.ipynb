```python
def investmentAmount(l, scale):
    return (15.0 * scale * scale + 200) * 1.0 / (1.0 * l / 10 + 1) + 200

#投資スキルのレベル
l = 140
#店の規模
scale = 61

for i in range(50):
    inv = investmentAmount(l+i, scale)
    print(l+i, inv)
```

```python
import random
#投資時に規模が2上がるか3上がるかはランダムと仮定

#region setting_const
whole_count_limit = 10000

#実際の投資の手順をシミュレーションする、ただし途中での投資レベル上昇を考えない
#roop_count_limit回の投資ごとに店員を雇い直すようにして、総投資回数分投資するまで続ける
#総投資回数を十分大きく取れば、店員を解雇するのを何回ごとにすればいいかが総投資回数に依存しにくいはず
def Go(roop_count_limit):
    random.seed(0)

    #region calc_variable

    #初期店規模もランダム性がある模様
    scale = random.randrange(13)

    count = 0

    roop_count = 0

    #whole investment amount
    whole_a = 0

    #region setting_variable

    #investment Level
    level = 154

    #実際の投資の手順を繰り返す
    while True:
        #temp variable
        #print(count, str(scale))
        a_now = investmentAmount(level, scale)
        whole_a += a_now
        #print(count, ":", "+" + str(a_now), ",", whole_a)
        count += 1
        roop_count += 1
        scale += 2 + random.randrange(2) # scale gains 2 or 3

        if count == whole_count_limit:
            break;

        #店員雇い直し
        if roop_count == roop_count_limit:
            scale = random.randrange(13)
            roop_count = 0
            whole_a += 20000
    
    #1回の投資あたりの金額
    print("★", whole_a / whole_count_limit, "★")

    
for i in range(25):
    print(str(i))
    Go(i)

    
```
